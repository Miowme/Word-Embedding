```python
def cleansing(df):
    df_clean=df.str.lower()
    df_clean=[re.sub(r"\d+","",i )for i in df_clean]
    df_clean=[re.sub(r'[^\w]', ' ', i)for i in df_clean]
    df_clean=[re.sub(r'\s+',' ',i)for i in df_clean]
    return df_clean
    
```

```python
clean_text=cleansing(df['text'])
```

```python
clean_text[0:2]
```

```python
for i in clean_text:
    print(i)
```

**Word2Vec - CBOW**

```python
import gensim
from gensim.models import Word2Vec 
```

```python
model_cbow = gensim.models.Word2Vec(word_token, min_count = 1,vector_size = 100, window = 5, sg=0)
#model_cbow = gensim.models.Word2Vec(word_token, min_count = 1,size = 100, window = 5, sg=0) for other version
```

```python
vocabulary_cbow = model_cbow.wv.index_to_key
print(vocabulary_cbow)
```

```python
print(model_cbow.wv.similarity('amazon','apple'))
```

```python
print(model_cbow.wv.similarity('analyst','buy'))
```

```python
similar_words = model_cbow.wv.most_similar("amazon")
print(similar_words)
```

```python
#get the vector
vectors_cbow = [model_cbow.wv.get_vector(word)for word in vocabulary_cbow]
```

```python
vectors_cbow[0]
```

**Fasttext**

```python
meta_hyper = {
    "vector_size": 4, # size of embedding 
    "alpha": 0.025, # learning rate
    "window": 5,
    "min_freq" : 1,
    "epochs": 3, # number of training epochs
    "vocab_size": 20, #len(set([td for td in word_token])), #size of vocabulary
    "data_description": "Add your description here",
    "data_size": len(word_token),
    "tokens_number": len([td for td in word_token]),
    "CPU": 8
}
```

```python
from gensim.models import FastText

embed_model = FastText(vector_size=meta_hyper['vector_size'], 
                       window=meta_hyper['window'], 
                       min_count = meta_hyper['min_freq'], 
                       alpha= meta_hyper['alpha'], 
                       workers=meta_hyper['CPU'])

embed_model.build_vocab(word_token)
```

```python
embed_model.train(word_token, total_examples=len(word_token), epochs=2)
```

```python
vector = embed_model.wv['amazon']
```
